```python
from typing import List, Tuple;
import os;

import numpy as np;

def _parseEmbCsv(uri: str) -> List[Tuple[str, str, np.ndarray]]:
    assert os.path.exists(uri);
    with open(uri, "r") as f:
        lines: List[str] = f.read().split("\n")[:-1];
    ret: List[Tuple[str, str, np.ndarray]] = [];
    _vecSize: int = len(lines[0].split(",")[2:]);
    lines = lines[1:];
    for l in lines:
        if l.__contains__('"'):
            _itm: List[str | float] = l.split('"');
            _code: str = _itm[0][:-1];
            _name: str = _itm[1];
            _itm = [float(_i) for _i in (_itm[2][2:]).split(",")];
            assert len(_itm) == _vecSize;
            ret.append((_code, _name, np.array(_itm, dtype=float)));
        else:
            _itm: List[str | float] = l.split(",");
            _code: str = _itm[0];
            _name: str = _itm[1];
            _itm = [float(_i) for _i in _itm[2:]];
            assert len(_itm) == _vecSize;
            ret.append((_code, _name, np.array(_itm, dtype=float)));
    return ret;
```

```python
l19: List[Tuple[str, str, np.ndarray]] = _parseEmbCsv("../data/icdEmb/icd-10-cm-2019-1000.csv");
l20: List[Tuple[str, str, np.ndarray]] = _parseEmbCsv("../data/icdEmb/icd-10-cm-2020-1000.csv");
l21: List[Tuple[str, str, np.ndarray]] = _parseEmbCsv("../data/icdEmb/icd-10-cm-2021-1000.csv");
l22: List[Tuple[str, str, np.ndarray]] = _parseEmbCsv("../data/icdEmb/icd-10-cm-2022-1000.csv");
```

```python
import pickle;
from typing import Dict;

tl: Dict[str, Tuple[str, np.ndarray]] = dict();
for _code, _name, _vec in l19:
    tl[_code] = (_name, _vec);
with open("d19.pkl", "wb") as f:
    pickle.dump(tl, f);

tl: Dict[str, Tuple[str, np.ndarray]] = dict();
for _code, _name, _vec in l20:
    tl[_code] = (_name, _vec);
with open("d20.pkl", "wb") as f:
    pickle.dump(tl, f);

tl: Dict[str, Tuple[str, np.ndarray]] = dict();
for _code, _name, _vec in l21:
    tl[_code] = (_name, _vec);
with open("d21.pkl", "wb") as f:
    pickle.dump(tl, f);

tl: Dict[str, Tuple[str, np.ndarray]] = dict();
for _code, _name, _vec in l22:
    tl[_code] = (_name, _vec);
with open("d22.pkl", "wb") as f:
    pickle.dump(tl, f);
```

```python
import os;
import pickle;
from typing import List, Tuple, Dict;

import numpy as np;

def _loadDict(pkl: str) -> dict:
    assert os.path.exists(pkl);
    with open(pkl, "rb") as f:
        return pickle.load(f);

dictList: List[Dict[str, Tuple[str, np.ndarray]]] = [_loadDict(_pkl) for _pkl in ["d19.pkl", "d20.pkl", "d21.pkl", "d22.pkl"]];

with open("allIcds.txt", "r") as f:
    allIcd: List[str] = f.read().split("\n")[:-1];

_icdHitMap: Dict[str, List[bool]] = dict();

for _icd in allIcd:
    _rcd: List[bool] = [False for _ in range(len(dictList))];
    for i in range(len(dictList)):
        try:
            dictList[i][_icd];
            _rcd[i] = True;
        except:
            pass;
    _icdHitMap[_icd] = _rcd;
```

```python
_ctNoMatch: int = 0;
_qualified: int = 0;
_qCode: Dict[str, int] = dict();
for k in list(_icdHitMap.keys()):
    # if k[0].isnumeric() or k[0] == "Z" or k[0] == "R":
    if k[0] == "Z" or k[0] == "R":
        continue;
    _qualified += 1;
    _qCode[k] = 0;
    if not _icdHitMap[k].__contains__(True):
        _ctNoMatch += 1;
        if _ctNoMatch <= 8:
            print(k)
            
print(f"{_ctNoMatch} no match, {_qualified} qualified codes, total {len(list(_icdHitMap.keys()))}")
```

```python
allPt: str = "../data/allPt.pkl";
assert os.path.exists(allPt);

with open(allPt, "rb") as f:
    ap = pickle.load(f);

import sys;
sys.path.append("..");

from obj.pt import *;

ptIcdStt: Dict[str, List[int]] = dict();
icdMissStt: Dict[str, int] = dict();
icdHitStt: Dict[str, int] = dict();
out: str = "allIcds.txt";
allIcds: List[str] = [];

with open("../data/kgEmb2.pkl", "rb") as f:
    icdMap = pickle.load(f)[0];

for _ptk in list(ap.keys()):
    pt: Pt = ap[_ptk];
    ptIcdStt[_ptk] = [0, 0];
    # print(pt.dem.sab, pt.dem.aYr, pt.dem.aMo, pt.dem.eth)
    for evt in pt.evtList:
        if evt.type != EvtClass.Dig:
            continue;
        # print(evt.time, evt.cont[0], end="");
        ptIcdStt[_ptk][0] += 1;
        allIcds.append(evt.cont[0]);
        try:
            icdMap[evt.cont[0]];
            try:
                icdHitStt[evt.cont[0]] += 1;
            except:
                if (evt.cont[0] == "W492"):
                    print("KGPD",evt.cont[0])
                icdHitStt[evt.cont[0]] = 1;
        except:
            try:
                assert _qCode[evt.cont[0]].__contains__(True)
                if (evt.cont[0] == "W492"):
                    print("YALE",evt.cont[0])
                try:
                    icdHitStt[evt.cont[0]] += 1;
                except:
                    icdHitStt[evt.cont[0]] = 1;
            except:
                ptIcdStt[_ptk][1] += 1;
                try:
                    icdMissStt[evt.cont[0]] += 1;
                except:
                    icdMissStt[evt.cont[0]] = 1;

ptIcdExtRate: List[float] = [];
for _ptk in list(ap.keys()):
    ptIcdExtRate.append(0 if ptIcdStt[_ptk][0] == 0 else ((ptIcdStt[_ptk][0] - ptIcdStt[_ptk][1]) / ptIcdStt[_ptk][0]));
ptIcdExtRate: np.ndarray = np.array(ptIcdExtRate);
print(f"Pt ICD hit rate: {np.quantile(ptIcdExtRate, (0, .25, .5, .75, .9, 1))}")

icdMiss2dig: Dict[str, int] = dict();
icdMissPrim: Dict[str, int] = dict();
for k in list(icdMissStt.keys()):
    try:
        icdMiss2dig[k[:2]] += icdMissStt[k];
    except:
        icdMiss2dig[k[:2]] = icdMissStt[k];
    try:
        icdMissPrim[k[0]] += icdMissStt[k];
    except:
        icdMissPrim[k[0]] = icdMissStt[k];

print(icdMissPrim)
```
